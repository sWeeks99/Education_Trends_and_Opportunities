```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress, pearsonr
import os
import seaborn as sns
```

```python
college_clean_df = pd.read_csv("../data/college_clean.csv")
college_clean_df.head()
```

```python
# Pie Graphs
pie1_data = college_clean_df['Private'].value_counts()
labels = "Private", "Public"
explode = (0.1, 0)
plt.figure(figsize=(5,5))
plt.pie(pie1_data, labels=labels, explode=explode, shadow=True, autopct='%1.1f%%')
plt.title ("Private vs Public Colleges")
```

```python
#based on Andrew's dataframe
schools = ['Private', 'Public']
avg_total_costs = [19091.0, 12355.0]

plt.figure(figsize=(8, 6))
bars = plt.bar(schools, avg_total_costs, color=['steelblue', 'darkorange'])

# Add labels to each bar with the corresponding value
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'${yval:.2f}', ha='center', va='bottom')

plt.xlabel('School Type')
plt.ylabel('Average Total Costs')
plt.title('Average Total Costs by School Type')
plt.show()
plt.savefig('../images/Average_Total_Costs_by_School_Type')
```
